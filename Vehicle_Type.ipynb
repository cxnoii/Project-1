```python
from scipy import stats
import scipy.stats as st
```

```python
# Plot the world average as a line chart
p_LA = plt.plot(Vehivle_type_LA["Year"], Vehivle_type_LA["passenger cars"], label="passenger cars")
light_LA = plt.plot(Vehivle_type_LA["Year"], Vehivle_type_LA["light trucks"], label="light trucks")
large_LA = plt.plot(Vehivle_type_LA["Year"], Vehivle_type_LA["large trucks"], label="large trucks")
m_LA = plt.plot(Vehivle_type_LA["Year"], Vehivle_type_LA["motocycles"], label="motorcycles")


plt.xlabel("Years")
plt.ylabel("Number of crashes")
plt.legend(loc="best")

plt.xticks(Vehivle_type_LA["Year"],rotation = 45)

plt.axvspan(2008, 2013, color='red', alpha=0.2)
plt.title("LA Vehicle Type")
# Print our chart to the screen

d_slope, d_int, d_r, d_p, d_std_err = stats.linregress(Vehivle_type_LA["Year"], Vehivle_type_LA["passenger cars"])
d_fit = d_slope * Vehivle_type_LA["Year"] + d_int
line_eq = "y = " + str(round(d_slope,2)) + "x + " + str(round(d_int,2))
plt.scatter(Vehivle_type_LA["Year"],Vehivle_type_LA["passenger cars"])
plt.plot(Vehivle_type_LA["Year"],d_fit,"--")
a = Vehivle_type_LA["Year"]
b = Vehivle_type_LA["passenger cars"]
#print(f"The r-squared is: {d_r**2}")
print(f"The correlation coefficient between year and passenger car in LA county is {round(st.pearsonr(a,b)[0],2)}")
plt.savefig("Vehicle Type LA.png")
plt.show()
```
