```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Vehivle_type_butte = pd.read_csv("Resources/Butte_Vehicle type.csv")
Vehivle_type_LA = pd.read_csv("Resources/LA_vehicle type.csv")
Vehivle_type_Orange = pd.read_csv("Resources/Orange_vehicle type.csv")
Vehivle_type_SF = pd.read_csv("Resources/SF_vehicle type.csv")
```

```python
# Plot the world average as a line chart
p = plt.plot(Vehivle_type_butte["Year"], Vehivle_type_butte["passenger cars"], label="passenger cars")
light = plt.plot(Vehivle_type_butte["Year"], Vehivle_type_butte["light trucks"], label="light trucks")
large = plt.plot(Vehivle_type_butte["Year"], Vehivle_type_butte["large trucks"], label="large trucks")
m = plt.plot(Vehivle_type_butte["Year"], Vehivle_type_butte["motocycles"], label="motorcycles")


plt.xlabel("Years")
plt.ylabel("Number of crashes")
plt.legend(loc="best")
plt.xticks(Vehivle_type_butte["Year"],rotation = 45)
plt.title("Butte County Vehicle Type")
plt.axvspan(2008, 2013, color='red', alpha=0.2)
# Print our chart to the screen
plt.savefig("Vehicle Type Plot/Vehicle Type Butte.png")

plt.show()
```
