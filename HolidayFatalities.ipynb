```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
holiday_fatalities_df = pd.read_csv("Resources\Alcohol-Impaired-Fatalities-During-Holidays.csv")
holiday_fatalities_df
```

```python
holiday_fatalities_df.dtypes
```

```python
year_list = ["2002", "2003", "2004", "2005", "2006", "2007", "2008",
            "2013", "2014", "2015", "2016", "2017", "2018", "2019"]
year_list_pre = ["2002", "2003", "2004", "2005", "2006", "2007", "2008"]
year_list_post = ["2013", "2014", "2015", "2016", "2017", "2018", "2019"]
holiday_list = ["New Years", "Memorial Day", "July 4th", "Labor Day", "Thanksgiving", "Christmas"]
```

```python
# Store data for each holiday scatterplot
new_years_fatalities_df = holiday_fatalities_df.iloc[:,1]
memorial_day_fatalities_df = holiday_fatalities_df.iloc[:,2]
independence_day_fatalities_df = holiday_fatalities_df.iloc[:,3]
labor_day_fatalities_df = holiday_fatalities_df.iloc[:,4]
thanksgiving_fatalities_df = holiday_fatalities_df.iloc[:,5]
christmas_fatalities_df = holiday_fatalities_df.iloc[:,6]
```

```python
# Plot Holiday Deaths Before & After Ridesharing
fig = plt.figure(figsize=(12,10))
plt.title("Impaired Driving Deaths - Holidays")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.scatter(year_list, new_years_fatalities_df)
plt.scatter(year_list, memorial_day_fatalities_df)
plt.scatter(year_list, independence_day_fatalities_df)
plt.scatter(year_list, labor_day_fatalities_df)
plt.scatter(year_list, thanksgiving_fatalities_df)
plt.scatter(year_list, christmas_fatalities_df)
plt.legend(holiday_list)
plt.show()
```

```python
# Plot New Years Holiday Deaths Before & After Ridesharing
plt.scatter(year_list, new_years_fatalities_df)
plt.title("Impaired Driving Deaths - New Years")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Holiday Deaths Before & After Ridesharing
plt.scatter(year_list, memorial_day_fatalities_df)
plt.title("Impaired Driving Deaths - Memorial Day")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Holiday Deaths Before & After Ridesharing
plt.scatter(year_list, independence_day_fatalities_df)
plt.title("Impaired Driving Deaths - July 4th")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Holiday Deaths Before & After Ridesharing
plt.scatter(year_list, labor_day_fatalities_df)
plt.title("Impaired Driving Deaths - Labor Day")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Holiday Deaths Before & After Ridesharing
plt.scatter(year_list, thanksgiving_fatalities_df)
plt.title("Impaired Driving Deaths - Thanksgiving")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Holiday Deaths Before & After Ridesharing
plt.scatter(year_list, christmas_fatalities_df)
plt.title("Impaired Driving Deaths - Christmas")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()
```
