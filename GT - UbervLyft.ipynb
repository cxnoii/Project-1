```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gt_uber = pd.read_csv("Resources/Google Trends - Uber.csv")
gt_uber = gt_uber.set_index(["Year"])
gt_uber
```

```python
countynames_list = gt_uber.columns.values.tolist()
rideshareyears  = ['2013', '2014', '2015', '2016', '2017', '2018', '2019']
rural = ['Butte/Placer/Shasta','Imperial']
urban = ['Los Angeles/Orange','San Diego','San Francisco']
```

```python
plt.plot(gt_uber[rural])
plt.xticks()
plt.title('Rural Google Trends for Uber (since Market Launch)')
plt.legend(rural,loc='best')
#set parameters for tick labels
plt.tick_params(axis='x', which='major', labelsize=7)
plt.tight_layout()
```

```python
plt.plot(gt_uber[urban])
plt.xticks()
plt.title('Urban Google Trends for Uber (since Market Launch)')
plt.legend(urban,loc='best')
#set parameters for tick labels
plt.tick_params(axis='x', which='major', labelsize=7)
plt.tight_layout()
```
