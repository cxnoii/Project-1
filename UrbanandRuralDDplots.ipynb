```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
county_dd_fatalities_df = pd.read_csv("Resources/California_Counties_Drunk_Driver_Fatalities.csv")
county_dd_fatalities_df = county_dd_fatalities_df.set_index(["Year"])
county_dd_fatalities_df
```

```python
countynames_list = county_dd_fatalities_df.columns.values.tolist()
prerideshare_year_list = ['2002', '2003', '2004', '2005', '2006', '2007', '2008']
rideshareyears  = ['2013', '2014', '2015', '2016', '2017', '2018', '2019']
rural = ['Butte','Imperial','Placer','Shasta']
urban = ['Los Angeles','Orange','San Diego','San Francisco']
```

```python
prerideshare_df = county_dd_fatalities_df.iloc[0:7]
```

```python
prerideshare_df
```

```python
rideshareyears_df = county_dd_fatalities_df.iloc[7:14]
rideshareyears_df
```

```python
plt.plot(prerideshare_df[rural])
plt.xlabel("Year")
plt.ylabel("Number of Fatalities")
plt.title('Rural Drunk Driving Fatalities Before Rideshare Apps')
plt.legend(rural,loc='best')
plt.tick_params(axis='x', which='major', labelsize=7)
plt.savefig('County Type DD Fatalities Plots/RuralDDpreridesharelineplot.png')
plt.tight_layout()
```

```python
plt.plot(prerideshare_df[urban])
plt.xlabel("Year")
plt.ylabel("Number of Fatalities")
plt.title('Urban Drunk Driving Fatalities Before Rideshare Apps')
plt.legend(urban,loc='best')
plt.tick_params(axis='x', which='major', labelsize=7)
plt.savefig('County Type DD Fatalities Plots/UrbanDDpreridesharelineplot.png')
plt.tight_layout()
```

```python
plt.plot(rideshareyears_df[rural])
plt.xlabel("Year")
plt.ylabel("Number of Fatalities")
plt.title('Rural Drunk Driving Fatalities After Rideshare App Release')
plt.legend(rural,loc='best')
plt.tick_params(axis='x', which='major', labelsize=7)
plt.savefig('County Type DD Fatalities Plots/RuralDDpostridesharelineplot.png')
plt.tight_layout()
```

```python
plt.plot(rideshareyears_df[urban])
plt.xlabel("Year")
plt.ylabel("Number of Fatalities")
plt.title('Urban Drunk Driving Fatalities After Rideshare App Release')
plt.legend(urban,loc='best')
plt.tick_params(axis='x', which='major', labelsize=7)
plt.savefig('County Type DD Fatalities Plots/UrbanDDpostridesharelineplot.png')
plt.tight_layout()
```
