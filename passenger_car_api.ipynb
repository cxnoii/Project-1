```python
county_fatals = []
county_numbers = [7,25,37,59,61,89,73,75]
county_list = ['Butte','Imperial','Los Angeles','Orange','Placer','Shasta','San Diego','San Francisco']

url = 'https://crashviewer.nhtsa.dot.gov/CrashAPI/crashes/GetCrashesByLocation?'
url_year = 'fromCaseYear=2014&toCaseYear=2014'
url_state = '&state=6'

for county in county_numbers:
        query1 = url + url_year + url_state + f'&county={county}&format=json' 
        response = requests.get(query1).json()
        cities = response['Results'][0]
        cities = pd.DataFrame(cities)
        county_fatals.append(cities['FATALS'].astype(int).sum())
```

```python
county_fatals_df = pd.DataFrame({'County' : county_list, '2014' : county_fatals})
```

```python
county_numbers = [7,25,37,59,61,89,73,75]

county_list = [
'Butte'
,'Imperial'
,'Los Angeles'
,'Orange'
,'Placer'
,'Shasta'
,'San Diego'
,'San Francisco'
]

year_list = [2013,2014,2015,2016,2017,2018,2019]

county_fatals_df = pd.DataFrame({'County' : county_list})

url = f'https://crashviewer.nhtsa.dot.gov/CrashAPI/crashes/GetCrashesByLocation?'
url_state = '&state=6'


for year in year_list:
    county_fatals = []
    for county in county_numbers:
        query1 = url + f'fromCaseYear={year}&toCaseYear={year}' + url_state + f'&county={county}&format=json' 
        response = requests.get(query1).json()
        cities = response['Results'][0]
        cities = pd.DataFrame(cities)
        county_fatals.append(cities['FATALS'].astype(int).sum())
    county_fatals_df[f'{year}'] = county_fatals
```
